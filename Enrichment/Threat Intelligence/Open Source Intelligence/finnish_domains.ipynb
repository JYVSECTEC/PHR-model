Load finnish (.fi) domains from [Traficom](https://www.traficom.fi/fi/viestinta/fi-verkkotunnukset)

The data can be queried like other REST APIs from the URL `https://odata.domain.fi/OpenDomainData.svc/Domains()`

Below is an example of getting the latest 100 granted domains in JSON format

```python
from IPython.display import JSON
import requests
url = "https://odata.domain.fi/OpenDomainData.svc/Domains()?$orderby=GrantDate desc&$top=100&$format=json"
r = requests.get(url)
JSON(r.json())
```

These values can be passed on to a Pandas DataFrame using the from_dict function from DataFrame class.

```python
import pandas as pd
data_dict = r.json()['value']
df = pd.DataFrame.from_dict(data_dict)
df.head(5)
```
